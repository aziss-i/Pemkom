## Tipe Data
terdapa 9 tipe data dalam python, yaitu:


*   **String** : Karakter. ciri " "/' '
*   **Integer** : Bilangan bulat
* **Float** : Bilangan desimal
* **Complex Number** : bilangan kompleks

* **Tuple** : himpunan tetap (tidak bisa di edit). ciri ()
* **Set** : himpunan elemen unik. ciri {}
* **Dictionary** : himpunan yang memiliki pasangan key dan value. ciri {}

### Operator Penugasan


```python
# = :
x=2
print("x = "+str(x))
print("x =",x)
```

```python
# += :
x+=2 #sama dengan x=x+2
```

```python
# -= :
x-=2 #sama dengan x=x-2
```

```python
# *= :
x*2 #sama dengan x=x*2
```

```python
# /=: :
x/=2 #sama dengan x=x/2
```

```python
# //=
x//2 #sama dengan x=x//2
```

```python
# %= :
x%=2 #sama dengan x=x%2
```

```python
# **= :
x**=2 #sama dengan x=x**2
```

### Operator logika

```python
# and :  bernilai benar jika kedua operan bernilai benar
print(3>2 and 2<3)
```

```python
# or : bernilai benar jika salah satu dari operan bernilai benar
print(3>2 or 2>3)
```

```python
# not : mengembalikan nilai
print(not False)
```

### Operator Identitas

```python
# is : memeriksa apakah operan kiri sama dengan operan kanan
x=10
y=5
x is y
```

```python
# is not :  memeriksa apakah operan kiri tidak sama dengan operan kanan
x=10
y=5
x is not y
```

### Operator keanggotaan

```python
# in : memeriksa apakah nilai tersebut termasuk anggota himpunan
3 in [1,2,3,4]
```

```python
# not in : memeriksa pakah nilai yang dicari tidak ada di himpunan
5 not in [1,2,3,4,5]
```

```python
"bagus" not in ["aku","dia","dan dirinya"]
```

```python
"bagus" in ["bagus ", "dan","ani"]
```

### operator bitwise

* | = or
